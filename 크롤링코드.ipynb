```python
import pandas as pd
from wordcloud import WordCloud
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('keyword.csv')
df.head()
```

```python
word_freq = dict(zip(df['키워드'], df['빈도수']))
```

```python
# 한글 폰트 경로 설정 (NanumGothicBold.ttf 파일이 존재해야 합니다)
font_path = '/usr/share/fonts/truetype/nanumfont/NanumGothicBold.ttf'  

# 워드클라우드 설정
plt.rcParams['font.family'] = 'ONE Mobile POP'
```

```python
def red_gradient_color(word, font_size, position, orientation, font_path, random_state):
    # 색상 값 생성 (0~255 범위)
    red = int(255 * (font_size / 100))
    green = int(30 * (font_size / 100))  # 글자 크기에 따라 초록색 비율 조정
    blue = int(30 * (font_size / 100))
    return f'rgba({red}, {green}, {blue}, 1)'  # RGBA 형식으로 반환
```

```python
wordcloud = WordCloud(font_path=font_path, 
                      width=1920,
                      height=1080,
                      background_color='white',
                      color_func=red_gradient_color).generate_from_frequencies(word_freq)
```

```python
plt.figure(figsize=(10, 5))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')  # 축 제거
wordcloud.to_file('wordcloud.png')
plt.show()
```
